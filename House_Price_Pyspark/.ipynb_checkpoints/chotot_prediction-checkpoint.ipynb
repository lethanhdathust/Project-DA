#### Check the percentage of null values and drop the columns has more than 51% percentages

```python
# [(count(when(col(c).isNull() | isnan(c), c)) * 100 / total_rows).alias(c) for c in columns]
columns = df_raw.columns
null_percentages = df_raw.select([(count(when(isnan(c) | col(c).isNull(), c)) *100 / df_raw.count()).alias(c) for c in columns]).collect()[0].asDict()
```

#### Fill null values

```python
df2.dtypes
```

Fill categorical values with mode values

```python
df3=df2
```

```python
# # Fill missing values using mode
for column in  [
'TongSoPhong',
'Huyen',
'GiayTo',
'LoaiHinhNhaO',
'PhongVeSinh',
'NoiThat',
'SoTang',
]:
    
    df_filtered = df2.filter(df2[column].isNotNull())
    mode_value = df_filtered.groupby(column).count().orderBy('count', ascending=False).first()[column]
    print(column, mode_value)
    df3 = df3.withColumn(column, when(col(column).isNull(), mode_value).otherwise(col(column)))
    # df3 = df3.withColumn(column, when(col(column) == 'null',mode_value).otherwise(col(column)))
```

```python

# # Fill missing values using mode
for column in  [
'ChieuNgang',
'ChieuDai',
'DienTich_m2'
]:
    # df_filtered = df2.filter(df2[column].isNotNull())
    # print(mean_value)
    # print(type(mean_value))
    mean_value = df2.select(avg(col(column))).first()[0]
    df3 = df3.withColumn(column, when(col(column).isNull(), mean_value).otherwise(col(column)))

df3.select('ChieuNgang',
'ChieuDai',
'DienTich_m2').show(100)
# ,''
```

#### Log Transform for Skewed Features¶


```python
for i,k in df_x.dtypes:
    print(i,k)
```

```python
# Select numeric columns
numeric_columns = [name for name, field in df_x.dtypes if field =='double' ]
# Calculate skewness for each numeric column
skewness_values = {col_name: df_x.select(skewness(col(col_name))).collect()[0][0] for col_name in numeric_columns}
# df_x.select(skewness(col('SoTang'))).collect()[0][0]
# # Convert the results to a pandas DataFrame
skew_df = pd.DataFrame(list(skewness_values.items()), columns=['Feature', 'skew'])
# # Calculate absolute skewness and determine if it's significantly skewed
skew_df['abs_skew'] = skew_df['skew'].abs()
skew_df['skewed'] = skew_df['abs_skew'] > 0.5
skew_df
```

Log transform each values

```python
for i in skew_df['Feature']:
   if skew_df.loc[skew_df['Feature'] == i ,'skewed'].values == True:
      df_x= df_x.withColumn(i , log1p(col(i) ))
```

# Model traning and testing

```python
merge_data = df5.join(k,how='inner')
merge_data
```

```python
assembler = VectorAssembler(inputCols=df5.columns.to_list(), outputCol="features")
spark_df =  py_spark.createDataFrame(merge_data)
assembled_df = assembler.transform(spark_df)
```

```python
spark_df.select('Gia').describe().show()
```

```python
# Select only the necessary columns
final_df = assembled_df.select("features", col('Gia').alias("label"))
# Split the data into training and test sets
train_data, test_data = final_df.randomSplit([0.8, 0.2], seed=100)
# Initialize the Linear Regression model
lr = LinearRegression(featuresCol="features", labelCol="label")
# Train the model
lr_model = lr.fit(train_data)
# Make predictions on the test data
predictions = lr_model.transform(test_data)

rf = RandomForestRegressor(featuresCol='features',labelCol='label')
rf_model = rf.fit(train_data)

predictions_rf = rf_model.transform(test_data)
```

```python
# Show predictions
predictions.select("features", "label", "prediction")
# Apply exponential function to convert log prices back to original prices
# predictions = predictions.withColumn("label_original", col("label"))
# predictions = predictions.withColumn("prediction_original", col("prediction"))
predictions.count()

```

```python
# Root Mean Squared Error (RMSE)
evaluator_rmse = RegressionEvaluator(predictionCol="prediction", labelCol="label", metricName="rmse")
rmse = evaluator_rmse.evaluate(predictions)
print(f"Root Mean Squared Error (RMSE) on original price data = {rmse}")

# Mean Absolute Error (MAE)
evaluator_mae = RegressionEvaluator(predictionCol="prediction", labelCol="label", metricName="mae")
mae = evaluator_mae.evaluate(predictions)
print(f"Mean Absolute Error (MAE) on original price data = {mae}")

# Mean Squared Error (MSE)
evaluator_mse = RegressionEvaluator(predictionCol="prediction", labelCol="label", metricName="mse")
mse = evaluator_mse.evaluate(predictions)
print(f"Mean Squared Error (MSE) on original price data = {mse}")

# R-squared (Coefficient of determination)
evaluator_r2 = RegressionEvaluator(predictionCol="prediction", labelCol="label", metricName="r2")
r2 = evaluator_r2.evaluate(predictions)
print(f"R-squared (R2) on original price data = {r2}")
```
